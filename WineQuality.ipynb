```python
#features and labels
x=next_df.iloc[:, :-1]
y=next_df['best_quality']
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=60)
```
